```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size' : 14})
```

```python
df = pd.read_csv('/home/pbr-student/Downloads/thesis-plots/mgn_to_eaone/compiled.csv')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)

fig.set_size_inches(10, 7)

var = ['epsilon', 'ca_eaone', 'ca_mgn']
demy = [0., 0.2, 0.5]

p = [None, None, None]

cbar_labels = ['%', '', '']

for row, v in enumerate( var ):
    pt = pd.pivot_table(df, values=v, index=['demy', 'growth_rate'], columns='update_interval')
    
    for col, d in enumerate( demy ):
        current_ax = ax[row][col]
        
        p[row] = current_ax.imshow( pt.loc[d] )
        
        current_ax.set_xticklabels([])
        current_ax.set_yticklabels([])

for r in range(3):
    cb = fig.colorbar(p[r], ax=ax[r].ravel().tolist(), shrink=0.95, label=cbar_labels[r])

for _ in range(3):
    ax[_][0].set_yticks([0, 1, 2, 3])
    ax[_][0].set_yticklabels([0.00001, 0.0001, 0.0005, 0.001])

for _ in range(3):
    ax[2][_].set_xticks([0, 1, 2])
    ax[2][_].set_xticklabels([1, 10, 50])
        
ax[0][0].set_ylabel("Connection Density\n\n");
ax[1][0].set_ylabel("eA1 Ca2+\n\nGrowth Rate (synaptic elements / ms)");
ax[2][0].set_ylabel("MGN Ca2+\n\n");

ax[2][1].set_xlabel("Update Interval (s)");

ax[0][0].set_title("0%");
ax[0][1].set_title("Percentage of MGN->eA1 Connections Demyelinated\n\n20%");
ax[0][2].set_title("50%");


```
