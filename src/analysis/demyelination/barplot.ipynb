```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size' : 14})
```

```python
# healthy, tonotopic, combined
# mgn, trn, eaone, iaone

mfr = np.array([
    [11.7, 7.8, 10.3, 18.7],
    [11.4, 7.5, 10.1, 18.4],
    [12.1, 8.7, 11.4, 16.5]
])

ifr_std = np.array([
    [4.1, 3.1, 2.7, 3.4],
    [8.5, 7.0, 3.4, 6.8],
    [9.0, 8.6, 5.8, 6.9]
])

pearson = np.array([
    [0.018, 0.009, 0.009, 0.004],
    [0.016, 0.018, 0.009, 0.003],
    [0.028, 0.138, 0.029, 0.129]
])

burst_gu = np.array([
    [13.6, 9.7, 16.9, 19.0],
    [10.9, 8.6, 15.7, 18.3],
    [13.4, 11.5, 20.0, 19.6]
])

burst_cerina = np.array([
    [0., 0., 0.4, 1.6],
    [0., 0., 0.3, 1.1],
    [0.2, 0.4, 3.3, 2.1]
])
```

```python
fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(12, 10))

fig.tight_layout()

populations = ['MGN', 'TRN', 'eA1', 'iA1']
colours = ['blanchedalmond', 'lightcyan', 'plum']
x = ['Healthy', 'Tonotopic', 'Combined']
labels = ['Healthy', 'Tonotopic', 'Combined']

for idx, population in enumerate(populations):
    ax[0][idx].bar(x, mfr[:, idx], color=colours)
    
    ax[1][idx].bar(x, burst_gu[:, idx], color=colours)
    ax[2][idx].bar(x, burst_cerina[:, idx], color=colours)
    
    ax[3][idx].bar(x, pearson[:, idx], color=colours)

for x in range(4):
    ax[0][x].set_title(populations[x])
    ax[3][x].set_xticklabels(labels, rotation=90)

for x in range(3):
    for y in range(4):
        ax[x][y].set_xticks([])

ax[0][0].set_ylabel("Mean Firing Rate (Hz)");
ax[1][0].set_ylabel("Burst Firing (%)\nGu et al. 2020");
ax[2][0].set_ylabel("Burst Firing (%)\nNarayanan et al. 2018");
ax[3][0].set_ylabel("Pearson Correlation\nCoefficient");

```
