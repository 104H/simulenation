### Summary

The weight of the connection gtom the MGN to the CTX increases as we go down the vertical axis. <br />
The left column shows results without tonotopicity, and the right column has tonotopicity.


As the weight of the connection from the MGN to the CTX increases,
1. In the tonotopic case, the activity in the MGN and the iA1 increases (dark colour to bright), but the activity in the eA1 decreases (bright colour to dark)
2. In the non-tonotopic case, the activity in the MGN and the iA1 stays similar (consistent dark colour), and the activity in the eA1 increases (dark colour to bright)
