```python
import nest
```

```python
MGN_params = {
        'E_L': -50.,  # resting membrane potential (mV) - see refs
        'C_m': 150.0,      # membrane capacity (pF)
        'g_L': 10.0,      # leak conductance  - see refs
        'V_reset': -55.,  # reset membrane potential after a spike (mV)  - for burstiness
        'V_th': -50.,  # spike threshold (mV)
        'tau_syn_ex': 2.5, # exc. synaptic time constant  - mit paper
        'tau_syn_in': 10., # exc. synaptic time constant  - mit paper

        "a": 40.,
        "b": 0.,
        'tau_w': 150.,
}


TRN_params = {
        'E_L': -60.,  # resting membrane potential (mV) - see refs
        'C_m': 150.0,      # membrane capacity (pF)
        'g_L': 10.0,      # leak conductance  - see refs
        'V_reset': -55.,  # reset membrane potential after a spike (mV)  - for burstiness
        'V_th': -50.,  # spike threshold (mV)
        'tau_syn_ex': 2.5, # exc. synaptic time constant  - mit paper
        'tau_syn_in': 10., # exc. synaptic time constant  - mit paper
        "a": 30.,
        "b": 80.,
        'tau_w': 300.
}
```

```python
nest.ResetKernel()

neuron = nest.Create("aeif_cond_exp", 1)

nest.SetStatus(neuron, MGN_params)

#neuron.V_m = -60. # (if needed)

pg_th = nest.Create('poisson_generator', n=1, params={'rate': 10})

nest.Connect(pg_th, neuron, 'all_to_all', syn_spec={'weight' : 100})

v = nest.Create("voltmeter")

nest.Connect(v, neuron)

nest.Simulate(1000.0)

nest.voltage_trace.from_device(v)
```
